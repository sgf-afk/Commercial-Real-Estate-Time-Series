Hotels would need their own set of unique data points and a larger data set that is less regionalized to be evaluated effectively. # Time Series Decomp & Analysis (annual data)

```python
annual_sqft = cleaned_df[['Date', 'idx_sqft_year']]
```

```python
series =  annual_sqft.set_index('Date')
decomposition = sm.tsa.seasonal_decompose(series, model='additive', freq = 365)

trend = decomposition.trend
trend =  trend[trend['idx_sqft_year'].isna()==False].reset_index()
trend.rename(columns = {'idx_sqft_year':'sqft_year_trend'}, inplace=True)
utils.ilinechart(trend, 'Date', 'sqft_year_trend', title = 'Trend')

seasonality = decomposition.seasonal.reset_index()
seasonality.rename(columns = {'idx_sqft_year':'sqft_year_seasonality'}, inplace=True)
utils.ilinechart(seasonality, 'Date', 'sqft_year_seasonality', title = 'Seasonality')

residuals = decomposition.resid.reset_index()
residuals.rename(columns = {'idx_sqft_year':'sqft_year_residuals'}, inplace=True)
utils.ilinechart(residuals, 'Date', 'sqft_year_residuals', title = 'Residuals')
```

```python
adf_test = sm.tsa.stattools.adfuller(series['idx_sqft_year'])

results = pd.Series(adf_test[0:4],
                   index = ['ADF Test Statistic', 'P-Value', 
                           '# Lags Used', '# Observations Used'])

for key,value in adf_test[4].items():
    results['Critical Value (%s)'%key] = value
    
print('Augmented Dickey-Fuller Test Results:\n')
print(results)
```

The annual sqft index is not stationary according to the Augmented Dickey-Fuller Test

# Time Series Decomp & Analysis (Daily data)

```python
annual_sqft = cleaned_df[['Date', 'idx_sqft_day']]
```

```python
series =  annual_sqft.set_index('Date')
decomposition = sm.tsa.seasonal_decompose(series, model='additive', freq = 365)

trend = decomposition.trend
trend =  trend[trend['idx_sqft_day'].isna()==False].reset_index()
trend.rename(columns = {'idx_sqft_day':'sqft_day_trend'}, inplace=True)
utils.ilinechart(trend, 'Date', 'sqft_day_trend', title = 'Trend')

seasonality = decomposition.seasonal.reset_index()
seasonality.rename(columns = {'idx_sqft_day':'sqft_day_seasonality'}, inplace=True)
utils.ilinechart(seasonality, 'Date', 'sqft_day_seasonality', title = 'Seasonality')

residuals = decomposition.resid.reset_index()
residuals.rename(columns = {'idx_sqft_day':'sqft_day_residuals'}, inplace=True)
utils.ilinechart(residuals, 'Date', 'sqft_day_residuals', title = 'Residuals')
```

Daily sqft index shows stationarity by rejecting the null hypothesis in the Augmented Dickey-Fuller test

```python
all_dates = pd.date_range('1970-10-26', periods = 18022, freq = 'D')
all_dates= pd.DataFrame(all_dates)

all_dates.rename(columns = {0:'Date'}, inplace=True)
# all_dates['Year'] = all_dates['Date'].dt.year
```

```python
rolling_avg_df['183_day'] = rolling_avg_df['idx_sqft_day'].rolling(window=183, min_periods=1).apply(lambda x : np.nanmean(x))
rolling_avg_df['365_day'] = rolling_avg_df['idx_sqft_day'].rolling(window=365, min_periods=1).apply(lambda x : np.nanmean(x))
rolling_avg_df['5_year'] = rolling_avg_df['idx_sqft_day'].rolling(window=1825, min_periods=1).apply(lambda x : np.nanmean(x))
```

```python
def wma(df, field, window):
    weights = np.arange(1, window + 1)
    ma = df[field].rolling(window)
    wma = ma.apply(lambda x: np.dot(x, weights)/weights.sum(), raw=True)
    return wma
```

```python
rolling_avg_df['183_day_wma'] = wma(rolling_avg_df, 'idx_sqft_day', 183)
rolling_avg_df['365_day_wma'] = wma(rolling_avg_df, 'idx_sqft_day', 365)
rolling_avg_df['5_year_wma'] = wma(rolling_avg_df, 'idx_sqft_day', 1825)
```

```python
model = sm.tsa.ExponentialSmoothing(rolling_avg_df['idx_sqft_day']).fit()
rolling_avg_df['365_day_SExp'] = model.predict(365)

dmodel = sm.tsa.ExponentialSmoothing(rolling_avg_df['idx_sqft_day'], trend = 'add').fit()
rolling_avg_df['365_day_DExp'] = dmodel.predict(365)

tmodel = sm.tsa.ExponentialSmoothing(rolling_avg_df['idx_sqft_day'], trend = 'add',
                                   seasonal='add',
                                   seasonal_periods = 4).fit()
rolling_avg_df['365_day_TExp'] = tmodel.predict(365)
```

```python
model = sm.tsa.ExponentialSmoothing(rolling_avg_df['idx_sqft_day']).fit()
rolling_avg_df['5_year_SExp'] = model.predict(1825)

dmodel = sm.tsa.ExponentialSmoothing(rolling_avg_df['idx_sqft_day'], trend = 'add').fit()
rolling_avg_df['5_year_DExp'] = dmodel.predict(1825)

tmodel = sm.tsa.ExponentialSmoothing(rolling_avg_df['idx_sqft_day'], trend = 'add',
                                   seasonal='add',
                                   seasonal_periods = 4).fit()
rolling_avg_df['5_year_TExp'] = tmodel.predict(1825)
```

```python
rolling_avg_df['SExp_Diff_183'] = rolling_avg_df['idx_sqft_day'] - rolling_avg_df['183_day_SExp']
rolling_avg_df['DExp_Diff_183'] = rolling_avg_df['idx_sqft_day'] - rolling_avg_df['183_day_DExp']
rolling_avg_df['TExp_Diff_183'] = rolling_avg_df['idx_sqft_day'] - rolling_avg_df['183_day_TExp']

rolling_avg_df['SExp_Diff_365'] = rolling_avg_df['idx_sqft_day'] - rolling_avg_df['365_day_SExp']
rolling_avg_df['DExp_Diff_365'] = rolling_avg_df['idx_sqft_day'] - rolling_avg_df['365_day_DExp']
rolling_avg_df['TExp_Diff_365'] = rolling_avg_df['idx_sqft_day'] - rolling_avg_df['365_day_TExp']

rolling_avg_df['SExp_Diff_5_year'] = rolling_avg_df['idx_sqft_day'] - rolling_avg_df['5_year_SExp']
rolling_avg_df['DExp_Diff_5_year'] = rolling_avg_df['idx_sqft_day'] - rolling_avg_df['5_year_DExp']
rolling_avg_df['TExp_Diff_5_year'] = rolling_avg_df['idx_sqft_day'] - rolling_avg_df['5_year_TExp']
```
